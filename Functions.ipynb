```python
def describe_student(name, age, course):
    print(f"My name is {name}. I am {age} years old. I am learning {course}")
```

```python
describe_student("Alice", 19, "Python") #positional arguments
```

```python
describe_student(course = "Python", name = "Alice", age=19) #keyword arguments
```

```python
#default values
def describe_student(name, age, course="Python"):
    print(f"My name is {name}. I am {age} years old. I am learning {course}")
```

```python
describe_student("James", 19)
```

```python
describe_student("James", 19, "SQL")
```

Print vs Return

```python
def sum_two_numbers(a, b):
    result = a + b
    print(result)
```

```python
sum_two_numbers(5, 2)
```

```python
first_result = sum_two_numbers(5, 2)
```

```python
print(first_result)
```

```python
def sum_two_numbers(a, b):
    result = a + b
    return result
```

```python
second_result = sum_two_numbers(2, 3)

print(second_result)
```

```python
third_result = sum_two_numbers(5, 6)

print(third_result)
```

Local vs Global variable

Any variable defined within a function is a local variable

```python
result = 8 #global variable
def sum_two_numbers(a, b):
    result = 5 #local variable
    result = a + b #local variable
    return result
```

```python
print(result)
```

Passing an arbitrary number of arguments

```python
def sum_of_numbers(*numbers):
    result = 0
    for num in numbers:
        result += num
    return result
```

```python
sum_of_numbers(1, 2)
```

```python
sum_of_numbers(1, 2, 10, 5)
```

```python
sum_of_numbers(1, 2, 10, 5, 6, 7, 8, 14, 35, 16)
```

## Anonymous Functions

```python
#to add two variables
add = lambda x,y: x + y
```

```python
print(add(2, 3))
```

```python
print(add(2, 6))
```

```python
power = lambda a,b:a ** b
```

```python
print(power(2,3))
```

```python
#using with the filter function

numbers = [1, 2, 3, 4, 5]
```

```python
#to filter a list to just even numbers

even_numbers = filter(lambda x: x % 2 == 0, numbers)
```

```python
print(list(even_numbers))
```

```python
odd_numbers = filter(lambda y: y % 2 == 1, numbers)
```

```python
print(list(odd_numbers))
```

```python
#using with the map function

print(numbers)
```

```python
#to find the squares of all elements in the list
squares = map(lambda x: x ** 2, numbers)
```

```python
print(list(squares))
```
